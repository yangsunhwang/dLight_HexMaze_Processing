```python
pd.set_option('display.max_columns', None)


def get_log_pchoos_v_costNben1(photrats,postInj=True):
    '''if postInj is true, only consider trials after injection,
    else, consider pre-injection time points.'''
    #print(photrats.triframe)
    if postInj :
        df = photrats.triframe[photrats.triframe["post-Inj"] == 1].copy()
    else :
        df = photrats.triframe[photrats.triframe["pre-Inj"] == 1].copy()
    #print(df)
    df['ratcodes'] = df.rat.astype('category').cat.codes
    seshs=df.session.unique()
    #print(seshs)
    for s in range(len(seshs)):
        sdf = df.loc[(df.session==seshs[s])&(df.tri>25)].copy()
        rdf = pd.DataFrame({'pRwdDif':photrats.get_lr_dif_val(\
            sdf,'nom_rwd'),'ldif':photrats.get_lr_dif_val(sdf,'dist')})
        rdf['choose_L'] = sdf.lrchoice.values
        #print(rdf['choose_L'])
        rdf['session']=s
        rdf['rat'] = sdf.ratcodes.values
        rdf['tri'] = sdf.tri.values
        rdf['block'] = sdf.block.values
        rdf['sex'] = sdf.sex.values
        rdf.loc[:,"rt-1"] = np.nan
        sdf.reset_index(inplace=True)
        #print(rdf)
        for p in range(3):
            lagInds = sdf.loc[sdf.port==p,:].index
            rdf.loc[lagInds,"rt-1"] = sdf.loc[lagInds,'rwd'].shift(1).values
        if s == 0:
            photrats.regdf = rdf
        else:
            photrats.regdf = photrats.regdf.append(rdf,ignore_index=True)
    photrats.regdf.loc[photrats.regdf.choose_L==2,'choose_L']=np.nan
    #print(photrats.regdf.head(50))
```

```python
def run_sklearnLogRegWithPval(X,y):
    lm = LogReg(fit_intercept=True).fit(X,y)
    params = np.append(lm.intercept_,lm.coef_)
    predictions = lm.predict(X)
    new_X = np.append(np.ones((len(X),1)), X, axis=1)
    M_S_E = (sum((y-predictions)**2))/(len(new_X)-len(new_X[0]))
    v_b = M_S_E*(np.linalg.inv(np.dot(new_X.T,new_X)).diagonal())
    s_b = np.sqrt(v_b)
    t_b = params/ s_b
    p_val =[2*(1-stats.t.cdf(np.abs(i),(len(new_X)-len(new_X[0])))) for i in t_b]
    return np.concatenate(params),p_val

def run_smLogRegWithPval(X,y):
    X = np.hstack([np.ones(len(X)).reshape(-1,1),X])
    mod = sm.Logit(y, X).fit()
    coefs = mod.summary2().tables[1]['Coef.'].values
    pvals = mod.summary2().tables[1]["P>|z|"].values
    return coefs,pvals

def add_scaledVarsToRegDff_prepost(photrats):
    photrats.regdf_pre.loc[:,"rwdDifScaled"] = valscale.fit_transform(\
        photrats.regdf_pre.loc[:,"pRwdDif"].values.reshape(-1,1))
    photrats.regdf_pre.loc[:,"lengthDifScaled"] = valscale.fit_transform(\
        photrats.regdf_pre.loc[:,"ldif"].values.reshape(-1,1))
    photrats.regdf_post.loc[:,"rwdDifScaled"] = valscale.fit_transform(\
        photrats.regdf_post.loc[:,"pRwdDif"].values.reshape(-1,1))
    photrats.regdf_post.loc[:,"lengthDifScaled"] = valscale.fit_transform(\
        photrats.regdf_post.loc[:,"ldif"].values.reshape(-1,1))
    
def calc_choiceRegWeightsByRat(photrats,a, b, c, d,e, f, postInj=True): #a sont les values a retirer
    
    print("choose left regDf not yet saved. Getting now...")
    get_log_pchoos_v_costNben1(photrats, postInj=postInj)
    add_scaledVarsToRegDff(photrats)
    #print(photrats.triframe.session.unique())
    regWeights = np.zeros((len(photrats.triframe.session.unique()),3))
    ratSigLevels = np.zeros((len(photrats.triframe.session.unique()),3))
    #print(regWeights,ratSigLevels)
    sessionslist = []
    for r in range(len(photrats.regdf.session.unique())):
        if r!=a and r!=b and r!=c and r!=d and r!=e and r!=f:
            print(r)
            #print(photrats.regdf.session.unique())
            regDf = photrats.regdf.loc[\
                (photrats.regdf.session==photrats.regdf.session.unique()[r])&(photrats.regdf.tri>25),\
                ["session","choose_L","rwdDifScaled","lengthDifScaled"]]
            regDf = regDf.loc[regDf.notnull().all(axis=1),:]
            y = regDf.loc[:,"choose_L"]
            print(y)
            X = regDf.loc[:,["rwdDifScaled","lengthDifScaled"]]
            print(X)                            
            betas,pvals = run_smLogRegWithPval(X,y)
            regWeights[r,:] = betas #coeff
            ratSigLevels[r,:] = pvals #p-values
            ratRegDf = pd.DataFrame(regWeights,columns=["intercept","relative p(R)","relative distance"])
            ratRegDf.loc[:,["intercept p value","relative p(R) p value","relative distance p value"]] = ratSigLevels
            
            sessions = photrats.triframe.session.unique()
            sex_values = photrats.triframe.groupby("session")["sex"].first().reindex(sessions).values
            ratRegDf["sex"] = sex_values
            sessionslist.append(r)
            
            if postInj :
                ratRegDf["postInj"] = 1
            else:
                ratRegDf["postInj"] = 0
    ratRegDf['session'] = sessions
    for r in range(len(photrats.regdf.session.unique())):
        if regWeights[r,:].all()==0:
            ratRegDf.drop([r], inplace=True)#supp les lignes 0
    
    return ratRegDf
```

```python
def caca(photrats):
    '''if postInj is true, only consider trials after injection,
    else, consider pre-injection time points.'''
    df = photrats.triframe.copy()
    df['ratcodes'] = df.rat.astype('category').cat.codes
    seshs=df.session.unique()
    #print(seshs)
    for s in range(len(seshs)):
        sdf = df.loc[(df.session==seshs[s])].copy()
        rdf = pd.DataFrame({'pRwdDif':photrats.get_lr_dif_val(\
            sdf,'nom_rwd'),'ldif':photrats.get_lr_dif_val(sdf,'dist')})
        rdf['choose_L'] = sdf.lrchoice.values
        #print(rdf['choose_L'])
        rdf['session']=s
        rdf['rat'] = sdf.ratcodes.values
        rdf['tri'] = sdf.tri.values
        rdf['block'] = sdf.block.values
        rdf['sex'] = sdf.sex.values
        rdf.loc[:,"rt-1"] = np.nan
        sdf.reset_index(inplace=True)
        #print(rdf)
        for p in range(3):
            lagInds = sdf.loc[sdf.port==p,:].index
            rdf.loc[lagInds,"rt-1"] = sdf.loc[lagInds,'rwd'].shift(1).values
        if s == 0:
            photrats.regdf = rdf
        else:
            photrats.regdf = photrats.regdf.append(rdf,ignore_index=True)
    photrats.regdf.loc[photrats.regdf.choose_L==2,'choose_L']=np.nan

```

```python
import statsmodels.formula.api as smf

def plotreglog(photrats,a, b, c, d,e): #a sont les values a retirer
    caca(photrats)
    
    
    plt.figure(figsize=(5, 5))
    ax = plt.axes()
    Predlist=[]
    for r in range(len(photrats.regdf.session.unique())):
        if r!=a and r!=b and r!=c and r!=d and r!=e:
            regDf = photrats.regdf.loc[\
                (photrats.regdf.session==photrats.regdf.session.unique()[r])&(photrats.regdf.tri>25),\
                ["session","choose_L","pRwdDif","ldif"]]
            regDf = regDf.loc[regDf.notnull().all(axis=1),:]
            y = regDf.loc[:,"choose_L"]
            #print(y)
            X = regDf.loc[:,["pRwdDif"]]
            print(X)                            
            
            mod = smf.logit(formula='choose_L~pRwdDif', data=regDf).fit()
            #mod = sm.Logit(y, X).fit()
            Frame = pd.DataFrame({'pRwdDif': np.linspace(-80, 80, 1000)})
            #Frame = sm.add_constant(Frame)  # Add a constant column for the intercept term
             
            Pred = mod.predict(Frame)
            Predlist.append(Pred)
            # Plot individual predictions
            ax.plot(Frame["pRwdDif"], Pred, color='black', alpha=0.4, linewidth=2)
            
    # Plot average prediction
    
    avg_pred = np.mean(Predlist, axis=0)
    ax.plot(Frame["pRwdDif"], avg_pred, color='red', linewidth=4)
    
    ax.set_xlabel("Relative value (p(reward) difference)",fontsize=50,fontweight='bold')
    ax.set_ylabel('P(Choose port)',fontsize=50,fontweight='bold')
    plt.plot()
```

```python
import scipy.stats as stats
#calculation of the average of the maximum peak per session
def average_peaks(photrats):
    AveragePeak = []
    NormalizedDf = pd.DataFrame(columns=["port","rwd", "pchosen","session","block","tri", "post-Inj","pre-Inj","green_z_scored", "session_type"])
    df=photrats.df
    for r in range(len(photrats.df.session.unique())):
        print(r)
        rwdInds = df.loc[(df.port.notnull())&(df.rwd==1)&(df.session==r+1)&(df['pre-Inj']==1)&(df.tri>25),:].index #prendre tous les indices
        #correspondants a un reward pour chaque session
        window = [0,250*2]
        rwdTraces = []
        for i in rwdInds:
            trace = df.loc[i+window[0]:i+window[1],'green_z_scored'].values
            if len(trace)==501:
                rwdTraces.append(trace)
        #print(rwdTracesmax)
        Mean = np.mean(rwdTraces, axis=0)
        Max=np.max(Mean)
        #print(Max)
        if Max>1: #inclusion criterion : z-score >1
            AveragePeak.append(Max) #on prend la moyenne de tous les max pour une session. Non ! on prend le max de la moyenne de toutes les traces pour une session
            #Average contient la moyenne de tous les max par session, donc liste de longueur len(session.unique())
            #print(AveragePeak)
            Normalized = photrats.df.loc[\
                    (photrats.df.session==photrats.df.session.unique()[r]),\
                    ["port","rwd", "pchosen", "session","block", "tri", "post-Inj","pre-Inj","green_z_scored","session_type"]]
            #Normalized = Normalized.loc[Normalized.notnull().all(axis=1),:]
            NormalizedDfbis = pd.DataFrame(Normalized,columns=["port","rwd", "pchosen", "session","block", "tri", "post-Inj","pre-Inj","green_z_scored", "session_type"])
            NormalizedDfbis.loc[:,"green_normalized"] = Normalized.loc[:,'green_z_scored']/AveragePeak[-1]
        
            NormalizedDf = pd.concat([NormalizedDf, NormalizedDfbis])
            #print(len(NormalizedDf[NormalizedDf['green_z_scored']>3]))
            #print(NormalizedDf["green_normalized"])
    
    return NormalizedDf
```

```python
from sklearn.linear_model import LinearRegression
def calc_DAselonPROB(photrats, postInj=True):
    signalnnfiltre = average_peaks(photrats)
    
    if postInj :
        signal = signalnnfiltre[signalnnfiltre["post-Inj"] == 1].copy()
    else :
        signal = signalnnfiltre[signalnnfiltre["pre-Inj"] == 1].copy()
    
    
    print(signal)
    beta1 = np.zeros((len(photrats.df.session.unique())+1,3))
    for r in signal.session.unique(): 
        #prendre tous les indices correspondants a un reward pour chaque session
        signalbis=signal[signal.session== r].copy()
        #print(signalbis)
        DAmean = []
        pchose = []
        print(signalbis.session_type)
        if signalbis.session_type.all()=='prob':
            rwdInds = signalbis.loc[(signalbis.port.notnull())&(signalbis.session==r)&(signalbis.tri>25),:].index 
        else:
            indices_block_1 = signalbis.loc[(signalbis.port.notnull())&(signalbis.session==r)&(signalbis['block'] == 1) & (signalbis['tri'] > 25), :].index
            indices_block_plus_1 = signalbis.loc[(signalbis.port.notnull())&(signalbis.session==r)&(signalbis['block'] > 1), :].index
            rwdInds = indices_block_1.union(indices_block_plus_1)
        window = [-250*2,0]
        for i in rwdInds:
            trace = signalbis.loc[i+window[0]:i+window[1],"green_normalized"].values 
            tracemean = np.mean(trace)
            #print(trace)
            if not pd.isna(signal.loc[i, "pchosen"]):
                DAmean.append(tracemean)
                pchose.append(signal.loc[i, "pchosen"])
        
        y = np.array(DAmean)
        print(r)
        print(y)
        x = np.array(pchose).reshape((-1, 1))
        print(x)
        # Fit the ordinary least squares (OLS) regression model
        model = LinearRegression().fit(x,y)
        
        r_sq = model.score(x, y)
        intercept = model.intercept_
        coeff = model.coef_
        beta1[r,:] = r_sq, intercept, coeff #coeff
        
        if coeff<-0.010:
            print(r)
    
    ratRegDf = pd.DataFrame(beta1,columns=["R2","intercept","coeffbeta1"])
    for r in range(len(photrats.df.session.unique())):
        if beta1[r,:].all()==0:
            ratRegDf.drop([r], inplace=True)#supp les lignes 0
    return ratRegDf.coeffbeta1
```

```python
def calc_Width(photrats, a, b, c, postInj=True):
    signalnnfiltre = average_peaks(photrats)
    if postInj :
        signal = signalnnfiltre[signalnnfiltre["post-Inj"] == 1].copy()
    else :
        signal = signalnnfiltre[signalnnfiltre["pre-Inj"] == 1].copy()
    print(signal)
    Width =[]
    r=0
    while r<len(signal.session.unique()): 
        if r!=a and r!=b and r!=c:
            tracelist = []
            DAmean = []
        
            #prendre tous les indices correspondants a un reward pour chaque session
            rwdInds = signal.loc[(signal.port.notnull())&(signal.rwd==1)&(signal.session==r+1)&(signal.tri>25),:].index 
            window = [0,250*2]
            for i in rwdInds:
                trace = signal.loc[i+window[0]:i+window[1],"green_normalized"].values 
                tracelist.append(trace) 
            
            DAmean = np.average(tracelist, axis = 0)
            
            DAmeanlist = DAmean.tolist()
           
            max_peak=np.max(DAmean)
            HalfMax = max_peak/2
            max_peak_index = DAmeanlist.index(max_peak)
            p1 = None
            p2 = None

            for i in range(max_peak_index, len(DAmeanlist)):
                if DAmeanlist[i] <= HalfMax:
                    p2 = i
                    break

            for i in range(max_peak_index, 0, -1):
                if DAmeanlist[i] <= HalfMax:
                    p1 = i
                    break
                    
            if p1 != None :
                if p2 != None :
                    print(r,p2,p1)
                    # Calculate the width of the peak response
                    width = p2 - p1
                    Width.append(width)

            
            r+=1
        else :
            r+=1
           
    return Width
```
